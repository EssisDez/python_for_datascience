## LINEAR REGRESSION

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create DataFrame
df = pd.DataFrame({"price": [56000, 89900, 100000], 
                  "area": [234, 876, 1000]})

# Create scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['area'], df['price'], marker='+', color="red")

# Add labels
plt.xlabel("Area")
plt.ylabel("Price")

# Display the plot
plt.show()
```

```python
#understand

from sklearn import linear_model
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[["area"]],df.price)
```

```python
#30/10/25
reg.predict([[600]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
800*reg.coef_ +reg.intercept_
```
