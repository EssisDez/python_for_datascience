```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80,85,90,95,100,105,110,115,120,125])
y = np.array([240,250,260,270,280,290,300,310,320,330])

plt.plot(x, y)
plt.xlabel("Average pulse")
plt.ylabel("Calories burned")
plt.title("Calories vs Average Pulse")  
plt.show()
```

```python
import matplotlib.pyplot as pyplot

pyplot.bar([0.25,2.25,3.25,5.25,7.25],[300,400,200,600,700],
        label="Carpenter", color="b", width = 0.5)
pyplot.bar([0.75,2.75,3.75,5.75,7.75],[50,30,20,50,60],
           label="Plumber", color="g", width = 0.5)
pyplot.legend()
pyplot.xlabel("days")
pyplot.ylabel("wage")
pyplot.title("Daily Wage Comparison")

pyplot.show()
```
